```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.mode.chained_assignment = None 
```

```python
df = pd.read_csv('StudentsPerformance.csv')
pd.set_option('display.max_columns',None)
df.head()
```

```python
df.describe()
```

```python
df["gender"].value_counts().plot.bar(color=['blue','red'],figsize=[3,3])
```

```python
df['race/ethnicity'].value_counts().plot.barh(color=['red','orange','yellow','green','blue'],figsize=(3,3))
plt.title("Frequency Distribution of Race/Ethnicity",family='Arial',fontweight='bold')
```

```python
df[['math score','reading score','writing score']].sum().plot.bar(color=['red','blue','green'],figsize=(3,3))
plt.title('Score',fontweight='bold')
```

```python
df['parental level of education'].value_counts().plot.bar(figsize=(3,3),color=['black','blue','red'],title='Frequency Distribution of Parental Level of Education')
```

```python
df['test preparation course'].value_counts().plot.bar(figsize=(3,3),color=['blue','red'],title='Test Preparation Course')
```

```python
fig , axes = plt.subplots(2,2,figsize=(20,10))
axes[0, 0].pie(df['gender'].value_counts(),
               labels=['Female', 'Male'],
               colors=['red', 'green'],
               autopct='%1.1f%%',
               shadow=True,)
axes[0, 0].set_title('Gender Distribution',
                     family='Arial',
                     fontweight='bold')

axes[0, 1].pie(df['race/ethnicity'].value_counts(),
               labels=['Group C', 'Group D','Group B','Group E','Group A'],
               colors=['red','orange','yellow','green','blue'],
               autopct='%1.1f%%',
               shadow=True,
               explode=[0.1,0,0.1,0,0])
axes[0, 1].set_title('Group',
                     family='Arial',
                     fontweight='bold')

axes[1,0].pie(df['parental level of education'].value_counts(),
               labels=['Some College', "Associate's Degree",'High School','Some High School',"Bachelor's Degree","Master's Degree"],
               colors=['red', 'green', 'blue', 'cyan','orange','grey'],
               autopct='%1.1f%%',
               shadow=True,
               )
axes[1,0].set_title('Parental Education', 
                     fontsize = 20,
                     family='Arial',
                     fontweight='bold')

axes[1, 1].pie(df[['math score','reading score','writing score']].sum(),
               labels=['math', 'reading','writing'],
               colors=['orange','yellow','blue'],
               autopct='%1.1f%%',
               shadow=True,
               )
axes[1, 1].set_title('Score',
                     family='Arial',
                     fontweight='bold')
```

```python
plt.figure(figsize=(20, 10))
sns.countplot(x='math score',data=df, palette = 'bright')
plt.title('Math Score Comparison',fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.countplot(x='reading score',data=df, palette = 'bright')
plt.title('Reading Score Comparison',fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.countplot(x='writing score',data=df, palette = 'bright')
plt.title('Writing Score Comparison',fontsize = 20)
plt.show()
```

```python
fig , axes = plt.subplots(1,3,figsize=(15,5))

axes[0].set_title('Math Scores')
axes[1].set_title('Reading Scores')
axes[2].set_title('Writing Scores')

sns.violinplot(y='math score',data=df,ax=axes[0],color='blue')
sns.violinplot(x='writing score',data=df,ax=axes[1],color='red')
sns.violinplot(y='reading score',data=df,ax=axes[2],color='green')
```

```python
fig , axes = plt.subplots(1,3,figsize=(15,5))

axes[0].set_title('Math Scores')
axes[1].set_title('Reading Scores')
axes[2].set_title('Writing Scores')

sns.barplot(x='gender',y='math score',data=df,ax=axes[0],palette='Set1',hue='gender',legend=True)
sns.barplot(x='gender',y='reading score',data=df,ax=axes[1],palette='Set1',hue='gender',legend=True)
sns.barplot(x='gender',y='writing score',data=df,ax=axes[2],palette='Set1',hue='gender',legend=True)
```
