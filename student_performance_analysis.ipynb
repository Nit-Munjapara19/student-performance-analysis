```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.mode.chained_assignment = None 
```

```python
df = pd.read_csv('StudentsPerformance.csv')
pd.set_option('display.max_columns',None)
df.head()
```

```python
df.describe()
```

```python
df["gender"].value_counts().plot.bar(color=['blue','red'],figsize=[3,3])
```

```python
df['race/ethnicity'].value_counts().plot.barh(color=['red','orange','yellow','green','blue'],figsize=(3,3))
plt.title("Frequency Distribution of Race/Ethnicity",family='Arial',fontweight='bold')
```

```python
df[['math score','reading score','writing score']].sum().plot.bar(color=['red','blue','green'],figsize=(3,3))
plt.title('Score',fontweight='bold')
```

```python
df['parental level of education'].value_counts().plot.bar(figsize=(3,3),color=['black','blue','red'],title='Frequency Distribution of Parental Level of Education')
```

```python
df['test preparation course'].value_counts().plot.bar(figsize=(3,3),color=['blue','red'],title='Test Preparation Course')
```
